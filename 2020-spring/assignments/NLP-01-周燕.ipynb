````python
def create_grammar(grammar_str, split='=>', line_split='\n'):
    grammar = {}
    i = 0
    dest = ''
    for line in grammar_str.split(line_split):
        if not line.strip(): continue
        # print(line)
        [exp,stmt] = line.split(split)
        i += 1
        if i==1:
            dest = exp
        #print(words)
        #exp, stmt=words[0],words[1]
        grammar[exp.strip()] = [s.split() for s in stmt.split('|')]
    return grammar,dest.strip()
````

````python
# test grammar understanding
gram,dest = create_grammar(tennis_commentary)
print(gram)
print(dest)
````

````python
# 随机产生一个句子
import random
def generate(gram, target):
    if target not in gram: return target # means target is a terminal expression
    
    expaned = [generate(gram, t) for t in random.choice(gram[target])]
    return ''.join([e if e != '/n' else '\n' for e in expaned if e != 'null'])
````

````python
generate(gram,dest)
````

````python
# 输入语法串，和需要生成的句子数量即可
def generate_n(grammar_str,n):
    gram,dest = create_grammar(grammar_str)
    #print(gram)
    #print(dest)
    sent_n = []
    for i in range(n):
        s = generate(gram,dest)
        sent_n.append(s)
        #print(generate(gram,dest))
    return sent_n
````

````python
# 产生新年祝福语
generate_n(BestWishes,10)
````

````python
# 澳网转播解说
generate_n(tennis_commentary,10)
````

````python
# 充分利用语料库，分词，统计，筛选，统计词频，二元词频

articles = df['comment'].tolist()
print(articles[:10])
````

````python
import re
def token(string):
    # we will learn the regular expression next course.
    return re.findall('\w+', string)
````

````python
print(token(articles[0]))
````

````python
import jieba
from collections import Counter
with_jieba_cut = Counter(jieba.cut(articles[1]))
print(with_jieba_cut)
````

````python
with_jieba_cut.most_common()[:10]
````

````python
# clean \w之外符号
articles_clean = [''.join(token(str(a)))for a in articles]
# 分词
tokens = []
i =0
for clause in articles_clean:
    tokens += (list(jieba.cut(clause)))
    i +=1
    if i % 100 ==0:
        print(i)
````

````python
wordscnt = Counter(tokens)
````

````python
print(len(wordscnt))
print(wordscnt.most_common()[:10])
````

````python
# 用可视化帮助理解词频和停用词
%matplotlib inline
import matplotlib.pyplot as plt
frequiences = [f for w, f in wordscnt.most_common(100)]
n =  [i for i in range(100)]
plt.plot(n, frequiences)
````

````python
import numpy as np
plt.plot(n, np.log(frequiences))
````

````python
def prob_1(word):
    return wordscnt[word] / len(tokens)
````

````python
print(prob_1('电影'))
````

````python
TOKEN_2_GRAM = [''.join(tokens[i:i+2]) for i in range(len(tokens[:-2]))]
print(TOKEN_2_GRAM[:10])
wordscnt_2 = Counter(TOKEN_2_GRAM)
````

````python
def prob_2(word1, word2):
    if word1 + word2 in wordscnt_2: return wordscnt_2[word1+word2] / len(TOKEN_2_GRAM)
    else:
        return 1 / len(TOKEN_2_GRAM)
````

````python
prob_2('我们', '在')
````

````python
def cut(string): return list(jieba.cut(string))
````

````python
def prob_s(s):
    words = cut(s)
    s_pro = 1
    
    for i, word in enumerate(words[:-1]):
        next_ = words[i+1]
        probability = prob_2(word, next_)
        s_pro *= probability    
    return s_pro
````

````python
N = 10
def generate_best(grammar_str, lang_model):
    sentences = generate_n(grammar_str,N)
    #print(sentences)
    probs = []
    for s in sentences:
        t = (s,lang_model(s))
        probs += [t] #prob_s
        # print(s)
        # print(probs)
    probs_sorted = sorted(probs,key=lambda x:x[1],reverse=True)
    return probs_sorted[0]
````
